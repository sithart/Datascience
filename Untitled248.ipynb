```python
x_dup=[]
y_dup=[]
for i in y1:
    if i<0:
        t=y1.index(i)
        t=t+1
        x_dup.append(t)
        y_dup.append(i)
```

```python
x_dup,y_dup
```

```python
x_dup=[]
y_dup=[]
i=0
while i<len(y1):
    if y1[i]<0:
        x_dup.append(x[i])
        y_dup.append(y1[i])
    i=i+1
```

```python
x_dup,y_dup
```
